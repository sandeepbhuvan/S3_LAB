```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Simple dictionary dataset
data_dict = {
    "Feature_1": [1, 2, 1, 8, 9, 8, 1.5, 8.5, 2.5, 9.5],
    "Feature_2": [1, 2, 3, 8, 9, 7, 2.5, 8.2, 3.5, 9.1],
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data_dict)

# Apply K-Means Clustering
n_clusters = 2  # Number of clusters
model = KMeans(n_clusters=n_clusters, random_state=42)

df["Clusters"] = model.fit_predict(df)
```

```python
# Cluster centers
centroids = model.cluster_centers_

# Visualizing the clusters
plt.figure(figsize=(8, 6))
plt.scatter(df["Feature_1"], df["Feature_2"], c=df["Clusters"], cmap="viridis", s=100, alpha=0.7, label="Cluster Points")
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", marker="X", s=200, label="Centroids")
plt.title(f"K-Means Clustering (n_clusters={n_clusters})", fontsize=16)
plt.xlabel("Feature 1", fontsize=14)
plt.ylabel("Feature 2", fontsize=14)
plt.legend()
plt.grid(True)
plt.show()

# Display the DataFrame with clusters
print("Clustered DataFrame:\n", df)
```
