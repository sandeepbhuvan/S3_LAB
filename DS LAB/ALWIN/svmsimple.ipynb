```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

# Generate synthetic dataset
data, labels = make_blobs(n_samples=300, centers=2, random_state=42, cluster_std=1.5)

# Train an SVM classifier
svm_model = SVC(kernel="linear", random_state=42)
svm_model.fit(data, labels)

# Plot the data points
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap="coolwarm", s=50, alpha=0.7)

# Get the coefficients and intercept for the decision boundary
coef = svm_model.coef_[0]
intercept = svm_model.intercept_[0]
x_vals = np.linspace(data[:, 0].min(), data[:, 0].max(), 100)
y_vals = -(coef[0] * x_vals + intercept) / coef[1]

# Plot the decision boundary
plt.plot(x_vals, y_vals, color="black", label="Decision Boundary")

# Highlight the support vectors
plt.scatter(
    svm_model.support_vectors_[:, 0],
    svm_model.support_vectors_[:, 1],
    s=100,
    facecolors="none",
    edgecolors="k",
    label="Support Vectors",
)

# Add title and legend
plt.title("SVM Decision Boundary", fontsize=14)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

# Generate synthetic dataset
X, y = make_blobs(n_samples=300, centers=2, random_state=42, cluster_std=1.5)

# Train an SVM classifier
s = SVC(kernel="linear", random_state=42)
s.fit(X, y)

# Plot the data points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="viridis", s=50, alpha=0.6)

# Get axis limits
ax = plt.gca()
xlimit = ax.set_xlim()
ylimit = ax.set_ylim()

# Create a meshgrid for decision boundary
xx, yy = np.meshgrid(np.linspace(xlimit[0], xlimit[1], 100),
                     np.linspace(ylimit[0], ylimit[1], 100))

# Calculate decision function for each point in the meshgrid
z = s.decision_function(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)

# Plot the decision boundary and margins
plt.contour(xx, yy, z, colors="k", levels=[-1, 0, 1], alpha=0.5, linestyles=["--", "-", "--"])

# Add labels and title
plt.title("SVM Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
